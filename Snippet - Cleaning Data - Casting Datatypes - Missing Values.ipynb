```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first" : ["Corey", "Jane", "John", 'Chris', np.nan, None, "NA"],
    "last" : ["Schafer", "Doe", "Doe", 'Schafer', np.nan, np.nan, 'Missing'],
    "email" : ["CoreyMSchafer@gmail.com", "JaneDoe@email.com", "JohnDoe@email.com", None, np.nan, 'Anonymous@email.com', 'NA'],
    "age" : ['33', '55', '63', '36', None, None, "Missing"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# Remove rows with no values
df.dropna()
```

```python
# Default dropna settings
df.dropna(axis='index', how='any')
```

```python
# Only drops rows will all values missing
df.dropna(axis='index', how='all')
```

```python
# drop columns with any missing values in this case drops all
df.dropna(axis='columns', how='any')
```

```python
# drop row if subset is missing

# keep rows with at least a subset filled in
# here how can be all or any, doesnt matter

df.dropna(subset=['email'])
```

```python
# drop if both of the subsets are missing in a row

df.dropna(how='all', subset=['last', 'email'])
```

```python
# dealing with customized missing values
# it depends on how we load the data

df = pd.DataFrame(people)
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(how='all', subset=['last', 'email'])
```

```python
# if I only want to see if there are missing values but not drop them:

df.isna()
```

```python
# Filling missing values with certain values:

df.fillna('MISSING')
```

```python
# Casting Datatypes
# here we have all strings or mix of things (object)

df.dtypes
```

```python
# we want age to be a number. if we ask for ,.mean() for example, we get error
# we need to convert it:

# interestingly np.nan is a float
type(np.nan)

# so we need to convert to float, because if we try to convert to int (error)
```
